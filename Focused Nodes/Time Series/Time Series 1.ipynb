To analyse time series using statistical methods we need stationarity to hold.

- Weak form stationarity
  - Constant mean
  - Constant variance
  - Consistent covariance between equal sized samples

- Strict form stationarity
  - Samples are from identical distribution

# Hypothesis tests

## Primer on hypothesis testing

We start with a hypothesis called __null hypothesis__($H_0$) which states the parameter we are testing is equal to zero. 
* Another way to think about this is results we found are due to chance.

Then we form an __alternative hypothesis__($H_A$) which states the parameter is different than zero.
* Another way to think about this is results we found are not due to chance.

After some math magic we are presented with a test statistic and if we are lucky, the p-value.

By the way the hypotheses are formed we are set out to disprove the null hypothesis.


__Terminology__:

* __Critical value__: The critical values represent the thresholds for the test statistic at certain significance level. It is picked from the table representing the distribution of the parameter.
* __Test statistic__: We get this by computing the test. We want this to be smaller than the critical value.
* __P value__: This answers the question "If the null hypothesis is true, what is the probability that we'd observe a more extreme test statistic in the direction of the alternative hypothesis than we did?".  We find this by computing the probability of test statistic under the distribition of the test statistic.
* __significance level__: The probability to reject the null hypothesis if it is correct. We set this to 1%, 5%, 10% before the experiment. This is also known as probability of commiting Type I error


We compare the __p-value__ with a significance level usually __0.05__ but lower is always preferred.

## Stationarity tests

Luckily there are some statistical tests defined for detecting non-stationarity.

What is a unit root?
   * Given an AR(1) process with equation given by $y_t = \phi*y_{t-1} + \epsilon$ we have a model where $\phi = 1$(Lies on the unit circle)
   * $E[y_t] = y_t$ and $Var[y_t ] = t\sigma^2$ which does depend on t, therefore not stationary.

### Dicky Fuller test

$H_0$: non-stationarity -> One lag auto correlation coefficent =  1

$H_1$: stationarity -> One lag auto correlation coefficent is < 1

The output of adfuller is read as follows:
- Test statistic
- P-value
- Number of lags used
- Number of observations
- Critical points
    * at 1%
    * at 5%
    * at 10%
- Best information criterion

```python
sm.tsa.adfuller(data["Temperature"])
```

```python
# We can write a function that shows us a clearer output

def stat_test(test_fn):
    test = test_fn
    if test[1] < 0.05:
        return f"We reject the null hypothesis with p-value {round(test[1], 4)}"
    else:
        return f"We fail reject the null hypothesis with p-value {round(test[1], 4)}"
```

```python
stat_test(sm.tsa.adfuller(data["Temperature"]))
```

- BIC := $k*log(n) - 2ln(\hat{L})$

# Autoregressive Model (AR)

An AR model is a model that uses lagged values to predict future values. Usually denoted by AR(n) where n is the lag period.

Example AR(1):
$$y_t = w + \phi y_{t-1} + \epsilon_{t}$$

For stability purposes $-1< \phi < 1$ needs to hold true holds true.

As you may noticed we don't have access to the epilon when we are making a prediction thus the above equation becomes
$$\hat{y}_t = w + \phi y_{t-1}$$ for prediction.

We can look at the pacf to decide on the number of lags. Number of non-zero partial autocorrelations hints at the order of an AR model.

Note: We can't rely on AR models for non-stationionary data.

# Partial Autocorrelation Function

```python
tsaplots.plot_pacf(data["Temperature"])
plt.show()
```

```python
# by looking at the plot we can see up to 10 lags are significant
model = sm.tsa.ARMA(data["Temperature"], order = (10,0))
results = model.fit()
print(results.summary())
```

```python
def plot_predict(model, data, steps):
    fig, ax = plt.subplots(figsize=(10, 5))
    model.plot_predict(start = len(data) - steps, end=len(data) + steps, ax = ax)
    plt.title("Forecast vs Actual")
    plt.show()
```

```python
plot_predict(results, data["Temperature"], 50)
```

# Moving Average Processes (MA)

MA tries to use the residuals/errors of past observations to model the time series. 
It is said to be shorter term.

Example MA(1):
$$y_t = \mu + \epsilon_{t} + \phi\epsilon_{t-1}$$

You can think of MA model as a model that uses the error/s it made in the past to deliver an accurate prediction.

We can look at the acf to decide on the number of lags.  Number of non-zero autocorrelations hints at the order of an AR model.


```python
# We can again look at the acf and this time figure out how many lags we need
tsaplots.plot_acf(data["Temperature"])
plt.show()
# It seems almost all lags are important, though we will later see how this is deceiving.
# For clarity I will only pick the highest 3.
```

```python
ma_model_res = sm.tsa.ARMA(data["Temperature"], order= (0,3)).fit()
ma_model_res.summary()
```

```python
plot_predict(ma_model_res, data["Temperature"], 50)
```

# Autoregressive Integrated Moving Average (ARIMA)

The itegrated part (I) refers to subtracting the value from 1 time step from the previous.
$$ y_t = y_t - y_{t-1} $$ The number of times we do this assingment is referred to as the order of the integration.

The differencing operation discussed above handles non-stationarity caused by unit root for us.

Explanation:
   * Suppose we have an AR(1) process whose equation is given by $y_t = \phi*y_{t-1} + \epsilon$
   * If we take the first difference we get $y_t - y_{t-1} = (\phi - 1)*y_{t-1} + \epsilon$
   * Remember since unit case refers to $\phi = 1$ then we have $y_t - y_{t-1} = \epsilon$
   * Then we have $E[y_t - y_{t-1}] = \epsilon$ and $Var[y_t - y_{t-1}] = \sigma^2$ which does not depend on t, therefore stationary.

Visually consider the following case:
   + We have an upwards sloping time series process

```python
def plot_diff(diff=False, return_df = False):
    date_range = pd.date_range(start='1/1/2018', end='1/08/2018', freq='H')
    gen = np.random.randint(0, 50 ,size = len(date_range))
    gen = [gen[i-1] * i/len(gen) for i in range(1,len(gen)+1)]
    df = pd.DataFrame(date_range, columns=['date'])
    df.set_index("date", inplace= True)
    df["data"] = gen
    
    if diff:
        df = df.diff(1).iloc[1:]
    
    if return_df:
        return df
    
    df.plot()
```

When we plot the series we can immedietly tell this is not a stationary process.

We can check this using th adfuller test.

```python
plot_diff()
plt.title(stat_test(sm.tsa.adfuller(plot_diff(return_df=True))))
plt.show()
```

The test verifies our assumption, this is indeed not a stationary process.

But when we difference it using diff() method, it resembles a stationary process.

We can again check this using the adfuller test.

```python
plot_diff()
plt.title(stat_test(sm.tsa.adfuller(plot_diff(diff=True, return_df=True))))
plt.show()
```

Now we have a stationary process using the I in ARIMA.

Careful considerations:
- Want to keep I as small as possible for a simpler model

# ARIMAX

In statististical modeling, the variable we are trying to explain with the model is called the __endogenous variable__, whereas the variables we use to explain the __endogenous variable__ is called the __exogenous varaible__.

In short: 
+ __endogenous variable__: explained within the model.   
+ __exogenous variable__: explained outside the model.

So far we have only included the endogenous variable "time" in our model.

We can also include exogenous variables for ARIMA models as long as they have predictive power.

# Seasonality revisited SARMA, SARIMA, SARIMAX

Adding seasonality to S(ARMA) S(ARIMA) S(ARIMAX)

When the data includes seasonality, it fundementally violates stationarity. Using S(ARIMA) we can remove the additive seasonal trends.

For __ARIMA__ models the order was given by:
+ p for the AR process
+ d for the integration(differencing)
+ q for the MA process
    
For __SARIMA__ models the order is given by:
+ P for the seasonal AR process
+ D for the seasonal integration(differencing)
+ Q for the seasonal MA process
+ s for the seasonal frequency
    
Thus the order representing a typical SARIMA process is given by (p,d,q),(P,D,Q,s)

Even though our data does not resemble a seasonal process, for the sake of demonstration we can assume it has a daily "seasonal" component.
(We have hourly data so daily would be 24 time steps )

```python
sarima_model = sm.tsa.SARIMAX(data["Temperature"], order= (2,1,1), seasonal_order = (1,1,1,5)).fit()
sarima_model.summary()
```

```python
pred = sarima_model.get_prediction(start=pd.to_datetime('1990-09-23'), dynamic=False)

ax = data["Temperature"]['1990-09-23':].plot(label='Actual')
pred.predicted_mean.plot(ax=ax, label='Prediction', alpha=.7)
plt.legend()




plt.show()
```
