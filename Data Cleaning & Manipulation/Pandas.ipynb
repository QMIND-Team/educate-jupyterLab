# Reading a csv file

Pandas read_csv function is flexible.

Some of the optional but useful arguments are:
```
sep : If you have a tab seperated file use '\t'
names : For you to pass columns names as a list.
usecols : To pick which columns you want to load.
dtype : To assign the datatypes on your own. Especially useful when dealing with large dataframes, can decrease memory requirements by downcasting.
nrows: How many rows you want.
```


````python
df = pd.read_csv("titanic.csv")
````

## Indexing

### Single Column

```df.column``` and ```df[column]``` are both correct.
```df.loc[]``` is for indexing with column names whereas ```df.iloc[]``` is for indexing with integers as you would a numpy array.

Note: ```df.loc``` and ```df.iloc``` return the underlying DataFrame not just a view.

````python
df.Survived.head()
````

````python
df["Survived"].head()
````

````python
# We can also return a dataframe with the extra brackets
df[["Survived"]].head()
````

## Multiple Columns

We can again use ```df.loc``` or ```df.iloc[]```.

````python
df[["Survived","Pclass"]].head()
````

## Conditional Indexing

````python
df[df["Pclass"] == 1]
````

## Multiple Conditional Indexing

When logical indexing with pandas, the original ```and``` is replaced with ```&``` and or is replaced by ```|```
<p>Furthermore, every condition should be enclosed with paranthesis.</p>

````python
df[(df["Pclass"] > 1) & (df["Survived"] == 1)]
````

````python
df[(df["Pclass"] > 1) | (df["Survived"] == 1)]
````

# Row Operations .apply
Pandas doesn't respond well to for loops.

.apply is a fast row operation.

It returns a view of a series and does not support the ```inplace = True``` argument.

Providing vectorized functions increases its efficiency.

````python
import numpy as np
````

````python
%%timeit
df.Age.apply(lambda x: x**2)
````

````python
df.Age.apply(lambda x: x**2)
````
