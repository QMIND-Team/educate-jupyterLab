## Basics <a class="anchor" id="figure"></a>

### The figure <a class="anchor" id="figure"></a>
The figure is the page that the graph is drawn on. The figure can be easily created as seen below.

```python
fig = plt.figure(num = 1,
                 #figsize = (4, 3), 
                 #dpi = 100.0, 
                 facecolor = 'white', 
                 edgecolor = 'black',
                 #frameon = True,
                 #clear = False,
                 #tight_layout = False,
                 #constrained_layout = False
                 )
xmin, xmax, ymin, ymax = [0, 10, 0, 10]
axes = plt.axis([xmin, xmax, ymin, ymax])
plt.xlabel("X - Label")
plt.ylabel("Y - Label")
plt.title("Title")

plt.show()
```

You can also have multiple plots in the same figure

```python
rows, cols = [2, 2]
fig, axes = plt.subplots(rows, cols)

(ax1, ax2), (ax3, ax4) = axes

ax1.set_title("Plot 1")
ax2.set_title("Plot 2")
ax3.set_title("Plot 3")
ax4.set_title("Plot 4")

fig.tight_layout(pad = 2)

plt.show()
```

### Customizing the axes <a class="anchor" id="custom-axes"></a>

```python
from matplotlib.ticker import MultipleLocator as mul
fig, axis = plt.subplots(figsize=(6, 6))

x = np.arange(0, 10, 0.1)
x2 = np.arange(0, 11, 1)
y = x2*2

axis.set_title('Title', size=20)
axis.set_xlabel("X - Label", size=15)
axis.set_ylabel("Y - Label", size=15)

axis.tick_params(axis='x', which='major', labelsize=15, rotation=0)
axis.xaxis.set_major_locator(mul(5))
axis.xaxis.set_major_formatter('{x:.02f}')
axis.xaxis.set_minor_locator(mul(0.5))

axis.tick_params(axis='y', which='major', labelsize=15, rotation=0)
axis.yaxis.set_major_locator(mul(5))
axis.yaxis.set_minor_locator(mul(0.5))

axis.set_aspect('equal') 
# You can also set the aspect ratio as a height/width ratio such as 2 to match the 'equal' ratio in this case

# You can also optionally turn of the axes if needed for a plot
# plt.axis('off')

axis.plot(x, x, color='blue', linestyle='dashed', linewidth=2)
axis.plot(x2, y, marker = 'o', linestyle=' ', color='green', markersize=10)

# The axes' limits can easily be set with
axis.set_xlim(0, 11)
axis.set_ylim(0, 22)

plt.show()
```

### Legend  <a class="anchor" id="legend"></a>

```python
x = np.arange(0, 2.1, 0.1)
y1 = x
y2 = x**2
y3 = x**3

fig, ax = plt.subplots()
plt.grid()
ax.plot(x, y1, 'g--', label='Linear')
ax.plot(x, y2, 'b-', label='Quadratic')
ax.plot(x, y3, 'r', label='Cubic')

ax.legend(loc='best', shadow=True, fontsize='x-large')

plt.show()
```

## Color Bar  <a class="anchor" id="color-bar"></a>

```python
fig, ax = plt.subplots()

x = np.arange(-5.1, 5.1, .1)  # len = 11
y = np.arange(-5.1, 5.1, .1)  # len = 7
x, y = np.meshgrid(x, y)
Z = -(x)**2 + (y)**2

ax.title.set_text("ColorMesh")
map = ax.pcolormesh(x, 
              y, 
              Z,
              alpha=1, 
              norm=None, 
              cmap='Blues', 
              edgecolors='none',
              vmin=None, 
              vmax=None, 
              shading='auto', 
              antialiased=True, 
              data=None)
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)

plt.colorbar(map, aspect=7)
plt.show()
```

## Plot Types <a class="anchor" id="plot-types"></a>

### Simple line plot <a class="anchor" id="simple-line-plot"></a>

```python
x = np.arange(0, math.pi*2, 0.05)
y = np.sin(x)

fig, ax = plt.subplots()

ax.plot(x, y)

ax.set_title("Title")
ax.set_xlabel("x label")
ax.set_ylabel("y label")

plt.show()
```

### Histogram <a class="anchor" id="histogram"></a>

```python
x = [1, 1, 1, 1, 2 ,2, 2 , 3, 3 ,4]

plt.hist(x, 
         bins=None, 
         range=None, 
         density=False, 
         weights=None, 
         cumulative=False, 
         bottom=None, 
         histtype='bar', 
         align='mid', 
         orientation='vertical', 
         rwidth=None, 
         log=False, 
         color=None, 
         label=None, 
         stacked=False,
         data=None)
plt.show()
```

### 2D Histogram <a class="anchor" id="2hist"></a>

```python
x = np.random.rand(100)
x = x.round(2)*3
y = np.random.rand(100)
y = y.round(2)*3
histogram = plt.hist2d(x, 
                      y, 
                      bins=10, 
                      range=None, 
                      density=False, 
                      weights=None, 
                      cmin=None, 
                      cmax=None, 
                      data=None)
plt.show()
```

### Bar chart <a class="anchor" id="bar"></a>

```python
x = [1, 2, 3, 4, 5]
y = [1, 2, 3, 4, 5]
fig, (one, two) = plt.subplots(1, 2)
one.title.set_text("Vertical")
two.title.set_text("Horizontal")
one.bar(x, 
      y, 
      width=0.8, 
      bottom=None, 
      align='center', 
      data=None)

two.barh(x, 
         y, 
         height=0.8, 
         left=None,
         align='center')
plt.show()
```

### Pie Chart <a class="anchor" id="pie"></a>

```python
values = [1, 2, 3, 4, 5]
labels = ['A', 'B', 'C', 'D', 'E']
explode = [0, 0, 0.1, 0, 0]

fig, axis = plt.subplots()
axis.title.set_text("Pie Chart")
axis.pie(values, 
         explode=explode,
         labels=labels, 
         #colors=colors,
         autopct=None,
         pctdistance=0.6, 
         shadow=True,
         labeldistance=1.1,
         startangle=0,
         radius=1,
         counterclock=True,
         wedgeprops=None, 
         textprops=None, 
         frame=False, 
         rotatelabels=False)
       
plt.show()
```

## Table <a class="anchor" id="table"></a>

```python
cellText=[['cell A1', 'cell B1', 'cell C1', 'cell D1'],
          ['cell A2', 'cell B2', 'cell C2', 'cell D2'],
          ['cell A3', 'cell B3', 'cell C3', 'cell D3'],
          ['cell A4', 'cell B4', 'cell C4', 'cell D4']       
         ]
rowLabels=[1, 2, 3, 4]
colLabels=['A', 'B', 'C', 'D']
x = [1, 2, 3, 4]
y = [1, 2, 3, 4]


fig, axis = plt.subplots()
axis.title.set_text("Table")

axis.bar(x, 
      y, 
      width=0.8, 
      bottom=None, 
      align='center', 
      data=None)

axis.table(cellText=cellText, 
           cellColours=None, 
           cellLoc='right', 
           colWidths=None, 
           rowLabels=rowLabels, 
           rowColours=None, 
           rowLoc='left', 
           colLabels=colLabels, 
           colColours=None, 
           colLoc='center', 
           loc='bottom', 
           bbox=None, 
           edges='closed')

plt.xticks([])
plt.yticks([0, 1, 2, 3, 4])

plt.show()
```

## Timeseries <a class="anchor" id="timeseries"></a>

```python
from pandas import read_csv
import matplotlib.dates as mdates
from matplotlib.dates import DateFormatter as df
import os

path = os.getcwd()

data = read_csv(path+'\SnP.csv', parse_dates=['Date'], index_col='Date', squeeze=True)


fig, axis = plt.subplots(3, 1, figsize=(15,15))

axis[0].plot(data['Open'])
axis[0].title.set_text('Price')
axis[0].xaxis.set_major_formatter(df("%Y"))
axis[0].tick_params(axis='x', which='major', labelsize=20, rotation=90)
axis[0].xaxis.set_minor_locator(mdates.YearLocator())
axis[0].xaxis.set_minor_formatter(df("%Y"))

axis[1].plot(data['Volume'])
axis[1].title.set_text('Volume')

axis[2].plot(data['2019']['Open'])
axis[2].title.set_text('2019')
axis[2].xaxis.set_major_formatter(df("%B"))
axis[2].tick_params(axis='x', which='major', labelsize=15, rotation=90)
axis[2].xaxis.set_major_locator(mdates.MonthLocator())
axis[2].xaxis.set_minor_formatter(df(""))
axis[2].xaxis.set_minor_locator(mdates.DayLocator(interval=2))

fig.tight_layout(pad=3)


plt.show()
```

## Polar plot <a class="anchor" id="polar"></a>

```python
th = np.arange(0, math.pi*2, 0.1)
r1 = np.sin(th)
r2 = np.cos(th)

plt.polar(th, r1, 'b')
plt.polar(th, r2, 'r')

plt.show()
```

## 3D Plotting <a class="anchor" id="3D"></a>

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})


theta = np.linspace(-4 * np.pi, 4 * np.pi, 100)
z = np.linspace(-2, 2, 100)
r = z**2 + 1
x = r * np.sin(theta)
y = r * np.cos(theta)

ax.plot(x, y, z)

plt.show()
```

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

x = np.random.rand(100)
y = np.random.rand(100)
z = np.random.rand(100)

ax.scatter(x, y, z, c=z, s=50)

ax.set_xlim(0,1)
ax.set_ylim(0,1)
ax.set_zlim(0,1)

plt.show()
```

## Stream plots <a class="anchor" id="stream"></a>

```python
x = np.arange(-10, 10, 0.1)
y = np.arange(-10, 10, 0.1)
x, y = np.meshgrid(x, y)
u = 1 - x**2
v = 1 - y**2

plt.streamplot(x, 
           y, 
           u, 
           v, 
           density=1, 
           linewidth=None, 
           color=(u+v)/2, 
           cmap=None, 
           norm=None, 
           arrowsize=1, 
           arrowstyle='-|>', 
           minlength=0.1, 
           transform=None, 
           zorder=None, 
           start_points=None, 
           maxlength=4.0, 
           integration_direction='both')

plt.show()
```
